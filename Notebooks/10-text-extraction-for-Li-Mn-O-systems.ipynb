```python
import os
# Load the pdfs one by one, make a dictionary with the pdf name as key and text as value
pdf_texts_dict = {}
folder_path = here("data/papers/Li-Mn-O")

for pdf_file_name in os.listdir(folder_path):
    # get file name without extension
    if not pdf_file_name.endswith(".pdf"):
        continue

    print(pdf_file_name)
    # Get the pdf name without the extension
    id = os.path.splitext(pdf_file_name)[0]
    print(f"Loading {id} ...")
    # Load the pdf text
    pdf_loader = PyMuPDFLoader(os.path.join(folder_path))
    pdf_text = pdf_loader.load_a_pdf(pdf_file_name)
    # Add the pdf text to the dictionary
    pdf_texts_dict[id] = pdf_text
# Write the pdf_texts_dict to a json file
import json
file_path = here("data/papers/Li-Mn-O/Li-Mn-O_dict.json")
with open(file_path, "w", encoding="utf-8") as f:
    json.dump(pdf_texts_dict, f, indent=4, ensure_ascii=False)


```
