### Select based on columns

```python
column_names = df.columns.to_list()
```

```python
interesting_topics = ["Product", "Composition", "Property", "Structure", "Space", "Article", "General Comments", "Warning"]
extracted_columns = []
for c in column_names:
    for i in interesting_topics:
        if c.find(i) >= 0:
            extracted_columns.append(c)

print(f"Extracted columns: {extracted_columns}")
```

```python
'Major Warning'.find("Warning")
```

```python
df = df[extracted_columns]
df.head()
```
