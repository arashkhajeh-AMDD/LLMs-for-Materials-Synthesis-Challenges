```python
from functools import reduce

# Suppose you have df1, df2, df3...
dfs = [ df_applications, df_comments_selected, df_melting_points, df_compositions[["ProductID"]+[col for col in df_compositions.columns if col.endswith("AtPercent")]]]

# Merge on 'ProductID'
merged_df = reduce(lambda left, right: pd.merge(left, right, on='ProductID', how='outer'), dfs)

merged_df
```

# Selecting Mn-Li-O systems

```python
import pandas as pd

# Your three required columns
required_atpercent_cols = ["Mn_AtPercent", "Li_AtPercent", "O_AtPercent", 'AllAtPercent']

# Identify all columns ending with 'AtPercent'
atpercent_cols = [col for col in merged_df.columns if col.endswith("AtPercent")]
print(atpercent_cols)

# Get the "other" AtPercent columns (not in the required list)
other_atpercent_cols = [col for col in atpercent_cols if col not in required_atpercent_cols]

# Filter
filtered_df = merged_df[
    merged_df[required_atpercent_cols].notna().all(axis=1) &  # required are not NaN
    merged_df[other_atpercent_cols].isna().all(axis=1)        # others are NaN
]

```

```python
filtered_df
```

```python
import ternary
import matplotlib.pyplot as plt

# Extract relevant atomic percent columns and drop NaNs
data = filtered_df[[ "Li_AtPercent", "O_AtPercent", "Mn_AtPercent",]].dropna()

# Normalize to 100% for ternary plot
data_normalized = data.div(data.sum(axis=1), axis=0) * 100

# Convert DataFrame rows to list of tuples
points = list(data_normalized.itertuples(index=False, name=None))

# Set up the ternary plot
figure, tax = ternary.figure(scale=100)
tax.set_title("Ternary Diagram: Mn-Na-O (Atomic Percent)", fontsize=15)

# Draw boundary and grid
tax.boundary()
tax.gridlines(color="gray", multiple=10)
tax.ticks(axis='lbr', linewidth=1, multiple=10, fontsize=10)

# Plot the points
tax.scatter(points, marker='o', color='blue', label="Compositions", s=30)

# Axis labels
tax.left_axis_label("Li", fontsize=12)
tax.right_axis_label("O", fontsize=12)
tax.bottom_axis_label("Mn", fontsize=12)

# Optional legend
tax.legend()

# Plot each point and annotate with element %s
for point in points[:5]:
    tax.scatter([point], marker='o', color='blue', s=30)
    label = f"Mn: {point[2]:.1f}%, Na: {point[0]:.1f}%, O: {point[1]:.1f}%"
    tax.annotate(label, position=point, fontsize=8)

# Show
plt.tight_layout()
tax.show()
```
