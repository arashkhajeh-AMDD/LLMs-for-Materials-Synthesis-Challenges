```python
import pickle
import pandas as pd
from pathlib import Path
from pyprojroot import here

table_name = "compositions_renamed.pickle"
data_dir = here("data")

file_path = data_dir / table_name

with open(file_path, "rb") as f:
    data = pickle.load(f)

df = pd.DataFrame(data)

df.head()
```

### Screening based on application

```python
# read application file

application_df = pd.read_csv(data_dir / "applications-energy.csv", dtype={'ProductID': str})
application_df.head()
```

```python
product_ids = application_df["ProductID"].to_list()

print(df[df.ProductID.isin(product_ids)])

```

```python
df_selected = df[df.ProductID.isin(product_ids)]
df_selected
```

```python
file_name = "compositions-energy.csv"

file_path = data_dir / file_name

df_selected.to_csv(file_path, index=False)
```
